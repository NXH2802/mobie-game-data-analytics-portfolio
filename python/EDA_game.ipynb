# Bước 6: Feature Engineering

```python
# Chuyển thời gian từ dạng string -> datetime
df['event_time_utc'] = pd.to_datetime(df['event_time_utc'])
```

```python
df["month"] = df["event_time_utc"].dt.month
```

```python
df["hour"] = df["event_time_utc"].dt.hour
```

```python
df["weekday"] = df["event_time_utc"].dt.day_name()
```

```python
# Cả 3 phiên bản có phát hành từ ngày nào? có song song không hay gối nhau
version_timeline = (
    df.groupby('app_version')
      .agg(
          first_seen=('event_time_utc', 'min'),
          last_seen=('event_time_utc', 'max'),
          users=('user_id', 'nunique')
      )
      .reset_index()
      .sort_values('first_seen')
)

version_timeline
```

```python
# Kiểm tra xem có người chơi nào, chơi ở 2 phiên bản khác nhau không, họ update giữa chừng.
multi_version_users = df.groupby('user_id')['app_version'].nunique()

multi_version_users[multi_version_users > 1]
```

```python
# Ngày cài game.
install_date = (
    df[df['event_name'] == 'first_open']
    .groupby('user_id')['event_time_utc']
    .min()
    .rename('install_date')
)

df = df.merge(
    install_date,
    on='user_id',
    how='left'
)
```

```python
# Ngày từ lúc tải game, phục vụ tính độ giữ chân (rêtention)
df['days_since_install'] = (
    df['event_time_utc'].dt.normalize()
    - df['install_date'].dt.normalize()
).dt.days
```

```python
# Người chơi có nạp tiền không
payer_map = (
    df.groupby('user_id')['iap_revenue_usd']
      .sum()
      .gt(0)
      .astype(int)
)

df['is_payer'] = df['user_id'].map(payer_map)
```
